**Вопрос 8:** (0.5 балл)

Какой вид активности SVM определяет хуже всего в терминах точности? Полноты? <br>

**Ответ:**
- по точности – подъем вверх по лестнице, по полноте – лежание
- по точности – лежание, по полноте – сидение
- по точности – ходьба, по полноте – ходьба
- по точности – сидение, по полноте – стояние

````python
import numpy as np


conf_matrix = np.array([
    [494, 2, 0, 0, 0, 0],  # True: ходьба (все 496)
    [12, 459, 0, 0, 0, 0],  # True: подъем (все 471)
    [2, 4, 413, 1, 0, 0],  # True: спуск (все 420)
    [0, 0, 0, 426, 65, 0],  # True: сидение (все 491)
    [0, 0, 0, 15, 517, 0],  # True: стояние (все 532)
    [0, 0, 0, 0, 11, 526]   # True: лежание (все 537)
])

# Вычисление precision: диагональ / сумма по столбцам (TP / (TP + FP))
precision = np.diag(conf_matrix) / np.sum(conf_matrix, axis=0)

# Вычисление recall: диагональ / сумма по строкам (TP / (TP + FN))
recall = np.diag(conf_matrix) / np.sum(conf_matrix, axis=1)

# Названия классов
classes = ['ходьба', 'подъем вверх по лестнице', 'спуск по лестнице', 'сидение', 'стояние', 'лежание']

# Вывод precision для каждого класса
print("Precision (точность) для каждого класса:")
for cls, prec in zip(classes, precision):
    print(f"{cls}: {prec:.3f}")

# Вывод recall для каждого класса
print("\nRecall (полнота) для каждого класса:")
for cls, rec in zip(classes, recall):
    print(f"{cls}: {rec:.3f}")

# Нахождение минимальных
min_prec_class = classes[np.argmin(precision)]
min_rec_class = classes[np.argmin(recall)]

print(f"\nКласс с минимальной precision: {min_prec_class} ({min(precision):.3f})")
print(f"Класс с минимальной recall: {min_rec_class} ({min(recall):.3f})")
````

**по точности – стояние, по полноте – сидение**
